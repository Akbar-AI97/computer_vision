```python
# Importing the OpenCV library
import cv2
```

```python
from google.colab.patches import cv2_imshow

# Reading the image using imread() function
image = cv2.imread('/content/cat_image.jpg')

cv2_imshow(image)
```

```python
print("Shape: ", image.shape) # returns tuple
# 483 => height
# 1024 => width
# 3 => BGR - Blue, Green, Red (color channels)

print("Height: ", image.shape[0])
print("Width: ", image.shape[1])
print("Color Channels: ", image.shape[2])

```

```python
(b, g, r) = image[100, 50] # pixel colors at 100th row and 50th column
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
# cvtColor() - Color Changing

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# resize() - Image Resizing

resized = cv2.resize(gray_image, (300, 300))
cv2_imshow(resized)
```

```python
# Image crop (cutting)

cropped = gray_image[80:400, 200:750] # image[y-axis, x-axis]
cv2_imshow(cropped)
```

```python
# rotate() Image rotation

rotated = cv2.rotate(cropped, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotated)
```

```python
# flip() - flip by y-axis

flipped = cv2.flip(cropped, 1)
cv2_imshow(flipped) # flipped

print("\n")
cv2_imshow(cropped) # original

```

```python
# Saving the edited image

cv2.imwrite("flipped_cat.jpg", flipped)
```

```python
# Image reading

flipped_cat = cv2.imread("/content/flipped_cat.jpg")
cv2_imshow(flipped_cat)
```
