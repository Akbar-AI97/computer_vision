```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Line Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Line Styles and Colors

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

# Multiple Lines in One Graph with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')
plt.title("Multiple Line Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

# Bar Chart with Numpy and Matplotlib

```python
x = np.array(["Food", "Expences", "Charity", "Salary"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Scatter Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.scatter(x, y, color='purple', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Histogram

```python
data = np.array([3, 4, 5, 6, 8, 8, 9, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Counts")
plt.show()
```

# Subplots - Multiple Graphs

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.figure(figsize=(8, 4)) #(width, height)

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red', marker='o')
plt.title("Subplot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue', linestyle='--')
plt.title("Subplot 2")

plt.tight_layout()
plt.show()
```

# Saving a plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(x, y, color='green')
plt.title("Save Figure Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.savefig("my_plot.png") # saves in current folder
plt.show()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
